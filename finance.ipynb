```python
from typing import List,Dict,Any
from toolz import curry
import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
def interest(amount,tax):
    return amount*(1+tax)
```

```python
# fixed income
@curry
def fixed_income(months : int, contribuition: float, fee: float , value: float) -> List[Any]:
    global array_values
    array_values = list()
    array_values.append(value)
    
    for _ in range(1,months): # 5 years
        value=interest(value+contribuition, fee) # 5% yield and R$ 500 increment (monthly)
        array_values.append(value)
    
    def plot(months: int, values : List)-> Any:
        fig = plt.figure(figsize=(12,8))
        plt.plot([m for m in range(1,months+1)], values ,color='g');
        plt.xlabel('Meses', {'weight':'bold', 'fontsize': 13 });
        plt.ylabel('Reais (R$)', {'weight':'bold', 'fontsize': 13  });
        plt.xticks( rotation='vertical' , fontsize=11, weight='bold');
        plt.yticks(fontsize=11, weight='bold');
        plt.title("Rendimento por tempo", fontsize=14, weight='bold');
        
        return fig
    
    
    return round(value,2), plot
```

```python
##### 50000
init_value = float(input("Valor inicial:"))
```

```python
#60
months = int(input("Em quantos meses:"))
```

```python
#500
monthly_contribution = float(input("Aporte mensal:"))
```

```python
#0.05
monthly_fee = float(input("Taxa mensal:"))
```

```python
fixed = fixed_income(months, monthly_contribution, monthly_fee)
```

```python
final_value , chart =  fixed(init_value)
```

```python
chart(months,array_values);
```

```python
# real estate funds
init_value=50000 
fund_price = 110
rentability = 0.85
months=240
```

```python
@curry
def estate_funds(months: int, fund_price: float,  rentability: float, init_value: float) -> List[Any]:
    global funds_array
    funds_array = list()
    funds_array.append(init_value)
    
    for _ in range(1,months):
        qtt = init_value//fund_price
        acumu = qtt*rentability
        init_value += acumu + 500
        funds_array.append(init_value)
        
    # function to plot
    def plot(months: int, values : List)-> Any:
        fig = plt.figure(figsize=(12,8))
        plt.plot([m for m in range(1,months+1)], values ,color='c');
        
        plt.xlabel('Meses', {'weight':'bold', 'fontsize': 13 });
        plt.ylabel('Reais (R$)', {'weight':'bold', 'fontsize': 13  });
        plt.xticks( rotation='vertical' , fontsize=11, weight='bold');
        plt.yticks(fontsize=11, weight='bold');
        plt.title("Rendimento por tempo", fontsize=14, weight='bold');
        
        return fig 
    
    return round(init_value,2), plot 
```

```python
funds = estate_funds(months, fund_price, rentability)
```

```python
final_value , chart = funds(init_value)
```

```python
final_rentab = (final_value//fund_price)*rentability
```

```python
chart(months,funds_array);
```
